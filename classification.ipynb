```python
from random_search import *
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
df = pd.read_csv('data/framingham.csv')
df = df.dropna()
```

```python
df_train = df.sample(frac=0.9)
df_valid = df.drop(df_train.index)
```

```python
X_train = df_train[['male','age','currentSmoker','cigsPerDay',
        'BPMeds','prevalentStroke']].to_numpy()
y_train = df_train['TenYearCHD'].to_numpy()

#,'education','prevalentHyp','diabetes', 'totChol','sysBP','diaBP', ,'age','currentSmoker','cigsPerDay',
       # 'BPMeds','prevalentStroke'


X_valid = df_valid[['male', 'age','currentSmoker','cigsPerDay',
        'BPMeds','prevalentStroke']].to_numpy()
y_valid = df_valid['TenYearCHD'].to_numpy()
```

```python
w, losses, valid_losses = random_search(300, X_train, y_train, X_valid, y_valid, 1e-5, 1e-6)
```

```python
plt.plot(list(range(300)), valid_losses, label = 'test')
plt.plot(list(range(300)), losses, label = 'train')
plt.legend()
```
